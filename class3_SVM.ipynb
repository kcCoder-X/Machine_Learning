```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the data
from sklearn.datasets import load_breast_cancer
cancer_Data = load_breast_cancer()
cancer_Data.keys()
```

```python
print(cancer_Data['DESCR'])
```

```python
cancer_Data['data']
```

```python
cancer_Data['feature_names']
```

```python
df_features = pd.DataFrame(cancer_Data['data'], columns=cancer_Data['feature_names'])
```

```python
df_features.isnull().sum()
```

```python
cancer_Data['target']
```

```python
cancer_Data['target_names']
```

```python
df_Target = pd.DataFrame(cancer_Data['target'], columns=['is_malignant_cancer'])
df_Target.tail(20)
```

```python
X = df_features
y = df_Target.values.ravel()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# model evaluation
from sklearn.metrics import classification_report, accuracy_score
print("Model accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# RUN VSM with"linear","C=10"

```python
from sklearn.metrics import accuracy_score

linear_svc = SVC(kernel='poly', C=10)
linear_svc.fit(X_train, y_train)

y_pred_linear = linear_svc.predict(X_test)

print("Linear model accuracy:", accuracy_score(y_test, y_pred_linear))
```

```python
linear_svc.score(X_train, y_train)
```

```python
linear_svc.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 2, 3, 4, 6, 10, 100],
               'kernel': ['linear', 'rbf', 'poly']}
opt_model= GridSearchCV(SVC(), param_grid)
opt_model.fit(X_train, y_train)
y_opt_pred = opt_model.predict(X_test)
```

```python
opt_model.best_params_
```

```python
opt_model.score(X_train, y_train)
```

```python
opt_model.score(X_test, y_test)
```
